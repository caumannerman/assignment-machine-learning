```python
import numpy as np

a = np.linspace(0,20,20)
b = np.linspace(0,30,30)

aa,bb = np.meshgrid(a,b)
```

```python
z = aa**2 + bb**2
```

```python
print(z.shape)
```

```python
import matplotlib.pyplot as plt
plt.contourf(aa,bb,z, 100, cmap="RdBu_r")
#plt.imshow(z, cmap = "RdBu_r")
plt.colorbar()
```

```python
plt.figure(figsize=(16,16))
#plt.contourf(aa,bb,z, 100, cmap="RdBu_r")
plt.imshow(z, cmap = "RdBu_r")
plt.colorbar()
```
