# Linear Regression 

## import library

```python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.mplot3d import Axes3D
```

## load point data for training and testing

```python
fname_data  = 'assignment_07_data.csv'

data        = np.genfromtxt(fname_data, delimiter=',')
num_data    = data.shape[0]

xx          = np.zeros(num_data)
yy          = np.zeros(num_data)
zz          = np.zeros(num_data)

for i in range(num_data):

    xx[i] = data[i,0]
    yy[i] = data[i,1]
    zz[i] = data[i,2]
print(xx.shape)
```

## plot the data in the three dimensional space

```python
fig = plt.figure(figsize=(12, 8))

ax1 = plt.subplot(111, projection='3d')
plt.title('data')

ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_zlabel('$z$')

ax1.scatter(xx, yy, zz, marker='o', color='blue', alpha=0.2)

plt.tight_layout()
plt.show()
```

## compute the loss function

```python
def compute_residual(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    residual = (theta[1]*x + theta[2]*y + theta[0]) - z
    num_data = xx.shape[0]
    residual = residual.reshape(2500,)
    # ============================
    return residual, num_data
```

```python
def compute_loss(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    residual, num_data = compute_residual(theta, x, y, z)
    loss = np.sum(residual ** 2) / (num_data*2)
    # ============================
    return loss
```

## compute the gradient for each model parameter (DO NOT COMPUTE THE GRADIENT FOR EACH MODEL PARAMETER, BUT DO COMPUTE THE GRADIENT OF THE MODEL PARAMTER VECTOR)

```python
def compute_gradient(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    #L(w) w  , 1/n ( X.T X w - X.T y ) 
    residual, num_data = compute_residual(theta,x,y,z)
    X = np.ones((2500,3))
    X[:,1] = x
    X[:,2] = y
    grad = (np.matmul(X.T.dot(X),theta.reshape(3,1)) - X.T.dot(z.reshape(z.shape[0],1))) / num_data
    grad = grad.reshape(3,)
    # ============================
    return grad
print(compute_gradient(np.array([0,0,0]), xx,yy,zz))
```

## gradient descent for each model parameter

```python
num_iteration   = 1000
learning_rate   = 0.01

theta           = np.array((0, 0, 0))
theta_iteration = np.zeros((num_iteration, theta.size))
loss_iteration  = np.zeros(num_iteration)

for i in range(num_iteration):
    # ===== FILL UP THE CODE =====
    theta = theta - compute_gradient(theta,xx,yy,zz)*learning_rate
    theta_iteration[i] = theta
    loss = compute_loss(theta,xx,yy,zz)
    loss_iteration[i] = loss

    # ============================
    print("iteration = %4d, loss = %5.5f" % (i, loss))
```

```python
f = theta[0] + theta[1] * xx + theta[2] * yy
```

## plot the results

```python
def plot_loss_curve(loss_iteration):

    plt.figure(figsize=(8,6))
    plt.title('loss')
    # ===== FILL UP THE CODE =====

    plt.plot(np.arange(1,1001), loss_iteration,'-', color = 'red')
    # ============================
    plt.tight_layout()
    plt.show()
```

```python
def plot_data(xx, yy, zz):
    
    fig = plt.figure(figsize=(12, 8))
    ax  = plt.subplot(111, projection='3d')
    plt.title('data')
    # ===== FILL UP THE CODE =====

    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')

    ax.scatter(xx, yy, zz, marker='o', color='blue', alpha=0.2)
    # ============================    
    plt.tight_layout()
    plt.show()
```

```python

def plot_model_parameter(theta_iteration):

    plt.figure(figsize=(8,6))
    plt.title('model parameter')
    # ===== FILL UP THE CODE =====
    plt.plot(np.arange(1,1001), theta_iteration[:,0],'-', color = 'red')
    plt.plot(np.arange(1,1001), theta_iteration[:,1],'-', color = 'green')
    plt.plot(np.arange(1,1001), theta_iteration[:,2],'-', color = 'blue')
    

    # ============================
    plt.tight_layout()
    plt.show()
```

```python
plot_loss_curve(loss_iteration)
```

```python
plot_model_parameter(theta_iteration)
```
